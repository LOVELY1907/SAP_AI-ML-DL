```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras import models,layers,regularizers
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.callbacks import EarlyStopping
```

```python
#data ingestion
data_path ='/content/sentiment_data_1000 (1).csv'
```

```python
#Tour of our Data
df = pd.read_csv(data_path)
df.head(3)
```

```python
#dependent and independent
x=df['text']
y=df['label']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from IPython.core.displayhook import tokenize
#tokenization
tokenizer= Tokenizer(num_words=5000,oov_token="<OOV>")
tokenizer.fit_on_texts(x_train)
```

```python
tokenizer.word_index
```

```python
vocab_size = len(tokenizer.word_index) + 1
print("vocabulary size:", vocab_size)
```

```python
#making sequence
x_train_seq = tokenizer.texts_to_sequences(x_train)
x_test_seq = tokenizer.texts_to_sequences(x_test)
```

```python
x_train_seq[0]
```

```python
#adding padding
x_train_padded = pad_sequences(x_train_seq,padding='post',maxlen=20)
x_test_padded = pad_sequences(x_test_seq,padding='post',maxlen=20)
```

```python
model1 = models.Sequential([
    #embedding layer
    layers.Embedding(input_dim=5000,output_dim=32,input_length=20),
    #RNN layer(hidden layer)
    layers.SimpleRNN(units=32,activation='tanh'),
    #output layer
    layers.Dense(1,activation='sigmoid')
])
```

```python
model1.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
history1 = model1.fit(
    x_train_padded,y_train,epochs=10,validation_data=(x_test_padded,y_test)
)
```

```python
sample_text = ['i feel not good']
sample_seq = tokenizer.texts_to_sequences(sample_text)
sample_pad = pad_sequences(sample_seq,maxlen=20,padding='post')
predictions =model1.predict(sample_pad)

for t, p in zip(sample_text, predictions):
  print(f"Text:{t} → Sentiment:{'Positive 😊' if p >0.3 else'Negative🤨'}{p}")
```

```python
model2 = models.Sequential([

    layers.Embedding(input_dim=5000,output_dim=64,input_length=20),
    layers.SpatialDropout1D(0.3),
    layers.SimpleRNN(units=64,activation='tanh',kernel_regularizer=regularizers.L2(0.001),dropout=0.3,recurrent_dropout=0.3),
    layers.Dense(32,activation='relu',kernel_regularizer=regularizers.L2(0.001)), # Closing parenthesis added here
    layers.Dropout(0.4),
    layers.Dense(1,activation='sigmoid')
])
```

```python
model2.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```

```python
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)
```

```python
history2 = model2.fit(
    x_train_padded,y_train,
    epochs=20,batch_size=32,
    validation_data=(x_test_padded,y_test),
    callbacks=[early_stop],
    verbose=1
)
```
