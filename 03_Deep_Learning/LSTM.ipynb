```python
faqs = faqs.lower()
faqs = faqs.translate(str.maketrans("", "", string.punctuation))
```

```python
#performing Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts([faqs])
```

```python
total_words = len(tokenizer.word_index) + 1
```

```python
total_words
```

```python
#creating sequences using tokens
input_sequences = []
words = faqs.split()
for i in range(1, len(words)):
    seq = words[:i+1]
    input_sequences.append(tokenizer.texts_to_sequences([' '.join(seq)])[0])
```

```python
max_len = max([len(x) for x in input_sequences])
input_sequences = np.array(pad_sequences(input_sequences, maxlen=max_len, padding='pre'))
```

```python
x, y = input_sequences[:, :-1], input_sequences[:, -1]
y = tf.keras.utils.to_categorical(y, num_classes=total_words)
```

```python
model = Sequential([
Embedding(total_words, 64, input_length=max_len - 1),
LSTM(128),
Dense(total_words, activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x, y, epochs=50, verbose=1)
```

```python
def predict_next_word(model, tokenizer, text, max_len):
    text = text.lower()
    token_list = tokenizer.texts_to_sequences([text])[0]
    token_list = pad_sequences([token_list], maxlen=max_len - 1, padding='pre')
    predicted = np.argmax(model.predict(token_list), axis=1)[0]
    for word, index in tokenizer.word_index.items():
        if index == predicted:
            return word
    return ""
```

```python
text = "The Codeunnati Program is"
next_word = predict_next_word(model, tokenizer, text, max_len)
print(f"Input: {text}\nPredicted next word: {next_word}")
```

```python
def generate_whole_text(model, tokenizer, text, max_len, num_words=20):
    text = text.lower()
    for _ in range(num_words):
        token_list = tokenizer.texts_to_sequences([text])[0]
        token_list = pad_sequences([token_list], maxlen=max_len - 1, padding='pre')
        predicted = np.argmax(model.predict(token_list), axis=1)[0]
        output_word = ""
        for word, index in tokenizer.word_index.items():
            if index == predicted:
                output_word = word
                break
        if output_word == "":
            break
        text += " " + output_word
    return text
```

```python
start_word = "The Codeunnati Program "
generated_line = generate_whole_text(model, tokenizer, start_word, max_len, num_words=30)
print(generated_line)
```
